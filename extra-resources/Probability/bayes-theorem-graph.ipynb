## Bayes theorem graph plotting

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def generate_likelihood_vs_posterior_for_priori(priori = 0.70):
    likelihood = np.linspace(0,1,101) # Probability of correctness
    posterior = (priori * likelihood) / (priori * likelihood + (1-priori)*(1-likelihood))
    return likelihood, posterior

a = 0.7
b, p = generate_likelihood_vs_posterior_for_priori(a)
```

```python
plt.plot(b,p)
```

```python
def draw_likelihood_vs_posterior_graph(likelihood, posterior, priori=None):
    plt.plot(likelihood, posterior)
    if priori is not None:
        plt.plot([.5, .5, 0], [0, priori, priori])
        plt.xticks([0,  .5, 1])
        plt.yticks([0, priori, 1])
        plt.xlabel(f'Marginal likelihood (Probability of rate at which guess is correct)\nGiven priori = {priori}')
        plt.ylabel('Posterior probability')

draw_likelihood_vs_posterior_graph(b, p, a)
```

```python
# animate
from IPython.display import clear_output
priories = np.linspace(0,1,101)
plt.draw()
for priori in priories:
    likelihoods, posteriors = generate_likelihood_vs_posterior_for_priori(priori)
    plt.cla()
    draw_likelihood_vs_posterior_graph(likelihoods, posteriors, priori)
    clear_output(wait=True)
    plt.show()
    
    
```
